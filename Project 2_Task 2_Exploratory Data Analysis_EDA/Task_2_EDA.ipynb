# ***<u>Identify Trends, Patterns and Visualization</u>***

### Plotting the distribution of the top 10 most frequent car makes

```python
top_10_makes = df['make'].value_counts().nlargest(10).index
df_top_10 = df[df['make'].isin(top_10_makes)]
plt.figure(figsize=(12, 6))
sns.countplot(data=df_top_10, y='make', order=top_10_makes, palette='viridis')
plt.title('Distribution of Top 10 Car Makes')
plt.xlabel('Number of Listings')
plt.ylabel('Car Make')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('top_10_makes_distribution.png')
plt.show()
```

### ***<u>Plotting the distribution of fuel types</u>***

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='fuel_type', palette='plasma')
plt.title('Distribution of Fuel Types')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Listings')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('fuel_type_distribution.png')
plt.show()
```

### ***<u>Plotting the distribution of transmission types</u>***

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='transmission', palette='magma')
plt.title('Distribution of Transmission Types')
plt.xlabel('Transmission Type')
plt.ylabel('Number of Listings')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('transmission_type_distribution.png')
plt.show()
```

### ***<u>Plotting a histogram for the price distribution</u>***

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True, color='blue')
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('price_distribution_histogram.png')
plt.show()
```

# <u>***Test Hypotheses and Validate Assumptions</u>***

### **Bivariate analysis: Price vs. Mileage**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='mileage', y='price', alpha=0.6)
plt.title('Price vs. Mileage')
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('price_vs_mileage_scatter.png')
plt.show()
```

### <u>***Bivariate analysis: Price vs. Year</u>***

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='year', y='price', alpha=0.6)
plt.title('Price vs. Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('price_vs_year_scatter.png')
plt.show()
```

### <u>***Calculate and print correlation coefficients</u>***

```python
correlation_mileage_price = df['mileage'].corr(df['price'])
correlation_year_price = df['year'].corr(df['price'])
print(f"\nCorrelation between Mileage and Price: {correlation_mileage_price:.2f}")
print(f"Correlation between Year and Price: {correlation_year_price:.2f}")
```

### <u>***Bivariate analysis: Price vs. Condition (handling missing values for this plot)</u>***

```python
df_clean_condition = df.dropna(subset=['condition'])
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_clean_condition, x='condition', y='price', palette='Paired')
plt.title('Price Distribution by Condition')
plt.xlabel('Condition')
plt.ylabel('Price')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('price_by_condition_boxplot.png')
plt.show()
```

### <u>***Bivariate analysis: Price vs. Seller Type</u>***

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='seller_type', y='price', palette='viridis')
plt.title('Price Distribution by Seller Type')
plt.xlabel('Seller Type')
plt.ylabel('Price')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('price_by_seller_type_boxplot.png')
plt.show()
```

# ***<u> Identify Trends, Patterns and Anomalies within the Dataset</U>***

1. #### Vehicle Characteristics:
- The most common car makes in the dataset are Toyota, Chevrolet, Volkswagen, Ford, and Honda. This suggests a strong market presence for these brands.
- In terms of fuel and transmission, Petrol and Diesel are the most common fuel types, and Automatic is the dominant transmission type, indicating a high demand for convenience and traditional engine types among used car buyers.
2. #### Relationships between Price, Year, and Mileage:
  As expected, a strong correlation exists between a car's price, year, and mileage:
   - Price vs. Mileage: There is a moderate negative correlation between price and mileage (
                              
                              r = -0.64
). This confirms the hypothesis that as mileage increases, the car's value tends to decrease.
- Price vs. Year: There is a strong positive correlation between price and year (

                              r = 0.75
                              
). This supports the idea that newer cars are generally more expensive than older ones.

3. #### Price Distribution by Condition and Seller Type
   The analysis shows a clear relationship between a car's condition and its price. On average, vehicles in 'like new' and 'excellent' conditions command significantly higher prices than those in 'good' or 'fair' conditions.

   When comparing prices by seller type, Certified Pre-Owned listings have the highest median price, followed by Dealer and Private Party listings. This suggests that buyers may be willing to pay a premium for the assurance and quality of a certified pre-owned vehicle.
